```
Temperature=int(input("Enter the temperature"))
Rain_condition=input("Enter if it's raining or not")
if Temperature>35 and Rain_condition=="Not raining":
  print("Stay hydrated")
elif (Rain_condition=="Raining" or Rain_condition=="raining") and Temperature<=35:
  print("Take an umbrella")
elif Temperature>35 and (Rain_condition=="Raining" or Rain_condition=="raining"):
  print("Stay hydrated and Take an umbrella")
else:
  print("Enjoy the weather")
```
